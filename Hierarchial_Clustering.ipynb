**Process**


1.  Every point is treaten as cluster
2.   The least distance points are joined upto last
3.   And a big clusters are formed at the final step.
4.  Every step is stored in a memory called Dendogram which is a graphical form of points and their euclidean distances



```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.cluster.hierarchy as sch
from sklearn.cluster import AgglomerativeClustering
```

```python
df= pd.read_csv('Mall_Customers.csv')
x=df.iloc[:,[3,4]].values
```

Dendogram is a plot between data points and eucledian distance in which

*    The least Distances are calculated until it reaches the 1 cluster
*  The no. of vertical lines next to a top horizontal line is taken as no. of clusters



```python
dendogram=sch.dendrogram(sch.linkage(x,method='ward'))
plt.title('Dendogram')
plt.xlabel('Customers')
plt.ylabel('Euclidean Distances')
plt.show()
```

No. of clusters we got are 5

```python
hc=AgglomerativeClustering(n_clusters=5,affinity='euclidean',linkage='ward')
y_hc=hc.fit_predict(x)
```
